```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('olympics.csv', index_col = 0, skiprows=1)
```

```python
for col in df.columns:
    if col[:2] == '01':
        df.rename(columns = {col:'Gold' + col[4:]}, inplace=True)
    if col[:2] == '02':
        df.rename(columns = {col:'Silver' + col[4:]}, inplace=True)
    if col[:2] == '03':
        df.rename(columns = {col:'Bronze' + col[4:]}, inplace=True)
    if col[:1] == '№':
        df.rename(columns = {col:'#' + col[1:]}, inplace=True)
```

```python
def answer_one():
    df.drop('Totals', inplace=True)
    return df['Gold'].idxmax()
answer_one()
```

```python
def answer_two():
    a = df['Gold'] - df['Gold.1']
    return a.idxmax()
answer_two()
```

```python
def answer_three():
    a = df[(df['Gold'] > 0) & (df['Gold.1'] > 0)  ]
    b = (a['Gold'] - a['Gold.1']) / a['Gold.2']
    return b.idxmax()
answer_three()
```

```python
#Q5

#def answer_five():

census_df = pd.read_csv('census.csv')
census_df.tail(180)
```

```python
#Q5

def answer_five():
    f = census_df['STNAME'].value_counts()
    return f.idxmax()
answer_five()
```

```python
#Q6
def answer_six():
    k = census_df.nlargest(3,'CENSUS2010POP')
    a = k['STNAME']
    return a.sort_index(axis = 0)
answer_six()
```

```python
# Q7

def answer_seven():
    z = census_df[['POPESTIMATE2010','POPESTIMATE2011','POPESTIMATE2012','POPESTIMATE2013','POPESTIMATE2014','POPESTIMATE2015']]
    return (z.max() - z.min()).abs().argmax()
answer_seven()
```

```python
#Q8
def answer_eight():
    r = census_df[((census_df['REGION'] == 1) | (census_df['REGION'] == 2)) & (census_df['CTYNAME'].str.startswith( 'Washington')) & (census_df['POPESTIMATE2015'] > census_df['POPESTIMATE2014'])  ]
    return r[['STNAME','CTYNAME']]
answer_eight()
```
